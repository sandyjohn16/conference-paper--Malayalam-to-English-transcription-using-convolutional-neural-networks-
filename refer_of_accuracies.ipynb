```python
import os
import librosa
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

# Function to calculate duration and amplitude from audio file
def calculate_duration_and_amplitude(audio_path):
    audio, _ = librosa.load(audio_path)
    duration = librosa.get_duration(y=audio)
    amplitude = np.max(np.abs(audio))
    return duration, amplitude

# Directory containing audio files
audio_dir = '/content/drive/MyDrive/voices'

# List to store duration, amplitude, and day of the week for each audio file
all_durations = []
all_amplitudes = []
all_days = []

# Iterate through audio files and calculate durations, amplitudes, and days of the week
for category in os.listdir(audio_dir):
    category_path = os.path.join(audio_dir, category)
    if os.path.isdir(category_path):
        for filename in os.listdir(category_path):
            if filename.endswith('.ogg'):
                audio_path = os.path.join(category_path, filename)
                duration, amplitude = calculate_duration_and_amplitude(audio_path)
                all_durations.append(duration)
                all_amplitudes.append(amplitude)
                all_days.append(category)  # Assuming the folder names are the days of the week

# Calculate mean, variance, skewness, and kurtosis for amplitude and duration for each day
days = sorted(set(all_days))
mean_amplitudes = [np.mean([amp for amp, day in zip(all_amplitudes, all_days) if day == d]) for d in days]
mean_durations = [np.mean([dur for dur, day in zip(all_durations, all_days) if day == d]) for d in days]
variance_amplitudes = [np.var([amp for amp, day in zip(all_amplitudes, all_days) if day == d]) for d in days]
variance_durations = [np.var([dur for dur, day in zip(all_durations, all_days) if day == d]) for d in days]
skewness_amplitudes = [skew([amp for amp, day in zip(all_amplitudes, all_days) if day == d]) for d in days]
skewness_durations = [skew([dur for dur, day in zip(all_durations, all_days) if day == d]) for d in days]
kurtosis_amplitudes = [kurtosis([amp for amp, day in zip(all_amplitudes, all_days) if day == d]) for d in days]
kurtosis_durations = [kurtosis([dur for dur, day in zip(all_durations, all_days) if day == d]) for d in days]

# Create a scatter plot
plt.figure(figsize=(7, 3))

# Plot mean amplitudes
sns.scatterplot(x=days, y=mean_amplitudes, s=150, color='blue', marker='o', label='Mean Amplitude')

# Plot mean durations
sns.scatterplot(x=days, y=mean_durations, s=150, color='orange', marker='s', label='Mean Duration')

# Plot variance of amplitudes
sns.scatterplot(x=days, y=variance_amplitudes, s=150, color='green', marker='^', label='Variance of Amplitude')

# Plot variance of durations
sns.scatterplot(x=days, y=variance_durations, s=150, color='red', marker='D', label='Variance of Duration')

# Plot skewness of amplitudes
sns.scatterplot(x=days, y=skewness_amplitudes, s=150, color='purple', marker='X', label='Skewness of Amplitude')

# Plot skewness of durations
sns.scatterplot(x=days, y=skewness_durations, s=150, color='brown', marker='P', label='Skewness of Duration')

plt.title('Statistics Across Days')
plt.xlabel('Days of the Week')
plt.ylabel('Value')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import os
import numpy as np
import soundfile as sf
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from sklearn.model_selection import StratifiedKFold

# Function to load data
def load_data(data_dir, max_length):
    data = []
    labels = []
    class_names = sorted(os.listdir(data_dir))

    for class_idx, class_name in enumerate(class_names):
        class_dir = os.path.join(data_dir, class_name)
        for audio_file in os.listdir(class_dir):
            if audio_file.endswith('.ogg'):
                audio_path = os.path.join(class_dir, audio_file)
                try:
                    # Load the audio file using soundfile
                    audio, _ = sf.read(audio_path, dtype='float32')

                    # Pad or truncate the audio to the specified max_length
                    if len(audio) < max_length:
                        padding = np.zeros(max_length - len(audio))
                        audio = np.concatenate((audio, padding))
                    else:
                        audio = audio[:max_length]

                    data.append(audio)
                    labels.append(class_idx)

                except Exception as e:
                    print(f"Error loading {audio_path}: {e}")

    return np.array(data), np.array(labels)

# Path to the data
data_dir = '/content/drive/MyDrive/voices'

# Parameters
sampling_rate = 44100
time_in_seconds = 2
max_length = int(time_in_seconds * sampling_rate)
num_classes = 7
batch_size = 32
epochs = 15

# Load data
data, labels = load_data(data_dir, max_length)

# Automatically determine input shape
input_shape = (data.shape[1], 1)  # Assuming data shape is (num_samples, sequence_length)

# Convert labels to categorical
labels_categorical = tf.keras.utils.to_categorical(labels, num_classes)
# Create a K-fold cross-validation iterator
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Iterate over folds
for fold_idx, (train_indices, val_indices) in enumerate(kfold.split(data, labels)):
    print(f'\nFold {fold_idx + 1}...')

    # Split data into training and validation sets
    train_data, val_data = data[train_indices], data[val_indices]
    train_labels, val_labels = labels_categorical[train_indices], labels_categorical[val_indices]

    # Build the model
    model = Sequential()
    model.add(Conv1D(filters=16, kernel_size=3, activation='relu', input_shape=input_shape))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Conv1D(filters=32, kernel_size=3, activation='relu'))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Flatten())
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))

    # Compile the model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

    # Train the model
    history = model.fit(train_data[..., np.newaxis], train_labels, epochs=epochs, batch_size=batch_size, validation_data=(val_data[..., np.newaxis], val_labels))

    # Evaluate the model on the validation set
    val_loss, val_accuracy = model.evaluate(val_data[..., np.newaxis], val_labels)
    print(f'Validation Loss: {val_loss:.4f}')
    print(f'Validation Accuracy: {val_accuracy:.4f}')
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score, recall_score, precision_score, f1_score, roc_curve
import matplotlib.pyplot as plt

# 1. Compute Predictions
val_predictions = model.predict(val_data[..., np.newaxis])

# 2. Compute Metrics
val_accuracy = accuracy_score(np.argmax(val_labels, axis=1), np.argmax(val_predictions, axis=1))
val_auc = roc_auc_score(val_labels, val_predictions, average='macro')
val_recall = recall_score(np.argmax(val_labels, axis=1), np.argmax(val_predictions, axis=1), average='macro')
val_precision = precision_score(np.argmax(val_labels, axis=1), np.argmax(val_predictions, axis=1), average='macro')
val_f1 = f1_score(np.argmax(val_labels, axis=1), np.argmax(val_predictions, axis=1), average='macro')

# 3. Visualize Results
# AUC Curve
fpr, tpr, thresholds = roc_curve(val_labels.ravel(), val_predictions.ravel())
plt.figure()
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % val_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()

# Print metrics
print(f'Validation Accuracy: {val_accuracy:.4f}')
print(f'Validation AUC: {val_auc:.4f}')
print(f'Validation Recall: {val_recall:.4f}')
print(f'Validation Precision: {val_precision:.4f}')
print(f'Validation F1 Score: {val_f1:.4f}')
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score, recall_score, precision_score, f1_score, roc_curve
import matplotlib.pyplot as plt

# 1. Compute Predictions
val_predictions = model.predict(val_data[..., np.newaxis])

# 2. Compute Metrics
val_accuracy = accuracy_score(np.argmax(val_labels, axis=1), np.argmax(val_predictions, axis=1))
val_auc = roc_auc_score(val_labels, val_predictions, average='macro')
val_recall = recall_score(np.argmax(val_labels, axis=1), np.argmax(val_predictions, axis=1), average='macro')
val_precision = precision_score(np.argmax(val_labels, axis=1), np.argmax(val_predictions, axis=1), average='macro')
val_f1 = f1_score(np.argmax(val_labels, axis=1), np.argmax(val_predictions, axis=1), average='macro')

# 3. Visualize Results
# AUC Curve
fpr, tpr, thresholds = roc_curve(val_labels.ravel(), val_predictions.ravel())
plt.figure()
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % val_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()

# Print metrics
print(f'Validation Accuracy: {val_accuracy:.4f}')
print(f'Validation AUC: {val_auc:.4f}')
print(f'Validation Recall: {val_recall:.4f}')
print(f'Validation Precision: {val_precision:.4f}')
print(f'Validation F1 Score: {val_f1:.4f}')
```
